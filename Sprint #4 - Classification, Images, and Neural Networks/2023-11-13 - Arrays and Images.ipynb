```python
im=imread('images/connect4.png')
```

```python
imshow(im)
```

```python
square_size=50
center_columns=arange(50,600,100)
center_columns
```

```python
imshow(im)
center_columns=arange(58,600,87)
for x in center_columns:
    plot([x,x],[0,530],'k-',lw=2)

center_rows=arange(58,580,87)
for y in center_rows:
    plot([0,600],[y,y],'g-',lw=2)
```

```python
center_rows
```

```python
center_columns
```

```python
locations=[]
for row in center_rows:
    for col in center_columns:
        locations.append([row,col])
locations
```

```python
imshow(im)
for r,c in locations:
    plot([c],[r],'o')
```

```python
r,c=[319, 58]
```

```python
r
```

```python
sr=r-square_size//2
er=sr+square_size
sc=c-square_size//2
ec=sc+square_size
sr,er,sc,ec
```

```python
subimage=im[sr:er,sc:ec,:]
subimage.shape
```

```python
imshow(subimage)
```

```python
count=1
for r,c in locations:
    sr=r-square_size//2
    er=sr+square_size
    sc=c-square_size//2
    ec=sc+square_size   
    subimage=im[sr:er,sc:ec,:]

    subplot(6,7,count)
    imshow(subimage)

    count+=1
```
